##Extract

extraindo dados da tabela vendas com Pandas


```python
import pandas as pd


df = pd.read_csv('vendas.csv', sep=';')
df

```

## Transform

Foram realizadas as seguintes alterações :

- Removido o 'R$" da coluna valor
- Alterada a virgula para ponto
- Realizada a conversão da coluna VALOR para tipo FLOAT64 para que pudessem ser realizados os calculos
- Adicionada a coluna TOTAL com base na multiplicação das colunas QUANTIDADE e VALOR

```python
df['VALOR'] = df['VALOR'].str.replace('R\$', '', regex=True)
```

```python
df   # O VALOR DA COLUNA TOTAL VAI DAR ERRADO DEVIDO AO TIPO DAS COLUNAS SEREM DIFERENTES
```

```python
df ['VALOR'] = df['VALOR'].apply(lambda x: str(x).replace(',','.'))
```

```python
df['VALOR'] = df['VALOR'].astype('float64')
```

```python
df['VALOR'] = df['VALOR'].astype('float64')
```

```python
df['TOTAL'] = df['VALOR'] * df['QUANTIDADE'] # O VALOR DA COLUNA TOTAL VAI DAR ERRADO DEVIDO AO TIPO DAS COLUNAS SEREM DIFERENTES
```
